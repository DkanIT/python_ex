# Exercise 7.2

```python
from collections import Counter
import collections
import re

file = open('shakespeare.txt', encoding="utf8")
a= file.read()
stopwords = set(line.strip() for line in open('shakespeare.txt'))
wordcount = {}
for word in a.lower().split():
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
# Print most common word
word_counter = collections.Counter(wordcount)
for word, count in word_counter.most_common(20):
    print(word, ": ", count)
# Close the file
file.close()
```

```python
uniques = []
for word, count in word_counter.most_common():
    if(count==1):
        uniques.append(word)
print(len(uniques))
# Close the file
file.close()
```

```python
for word, count in word_counter.most_common():
    if(count>=5):
        print(word, ": ", count)
```

```python
with open('common200.txt', 'w') as f:
    for word, count in word_counter.most_common(200):
        f.write('%s : %s \n' %(word, count))
```

```python
import random


def randomInt(n,a,b,filename):
    file = open(filename,"w")
    intList = [random.randrange(a,b)for n in range(n)]
    #print(intList)
    for i in intList:
        file.write("%s\n" % i)
    file.close()

randomInt(10,1,10,"randomInt.txt")

```

```python
def readrandomInt(filename):
    file = open(filename,'r')
    for val in file:
        print(val)
readrandomInt("randomInt.txt")
file.close()
```

```python
randomInt(10,1,10,"randomInt2.txt")
def sumofFiles(k,filename1,filename2):
    file1 = open(filename1,"r")
    file2 = open(filename2,"r")
    list1 = []
    list2 = []
    sum_list = []
    for item in file1:
        list1.append(int(item.strip()))

    for item in file2:
        list2.append(int(item.strip()))
    for item1 in list1:
        for item2 in list2:
            if(item1+item2==k):
                sum_list.append("u:%s and v:%s " %(item1,item2))
                
    for i in sum_list:
        print(i)


sumofFiles(10,"randomInt.txt","randomInt2.txt")
```
