```python
from collections import Counter
import collections
import re

file = open('shakespeare.txt')
a= file.read()

#A part

stopwords = set(line.strip() for line in open('shakespeare.txt'))
countedwords = {}
uniquewords = []
wordsatleast5times = []
for word in a.lower().split():    
    if word not in stopwords:
        if word not in countedwords:
            countedwords[word] = 1
        else:
            countedwords[word] += 1
word_counter = collections.Counter(countedwords)
for word, count in word_counter.most_common(20):
    print(word, ": ", count)

#B part
for word, count in word_counter.most_common():
    if(count==1):
        uniquewords.append(word)
print("--------------------------------------")
print("Unique words number : ",len(uniquewords))
print("--------------------------------------")

#C part
for word, count in word_counter.most_common():
    if(count>=5):
        wordsatleast5times.append(word)
        
print("Number of words which using at least 5 times : ",len(wordsatleast5times))
print("--------------------------------------")
#D part
c = open("mostcommon200.txt","w")

word_counter = collections.Counter(countedwords)
for word, count in word_counter.most_common(200):
    c.write("%s : %s \n" %(word,count))
print("Printed most common 200 words")
c.close()
```
